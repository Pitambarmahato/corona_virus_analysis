```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('corona_preprocess.csv')
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df['TotalCases'] = df['TotalCases'].str.replace(',', '')
df['TotalDeaths'] = df['TotalDeaths'].str.replace(',', '')
df['TotalRecovered'] = df['TotalRecovered'].str.replace(',', '')
df['Serious, Critical'] = df['Serious,Critical'].str.replace(',', '')
```

```python
df.info()
```

```python
# df['NewCases'] = df['NewCases'].astype(float)
df['TotalCases'] = df['TotalCases'].astype(float)
df['Serious, Critical'] = df['Serious, Critical'].astype(float)
df['TotalDeaths'] = df['TotalDeaths'].astype(float)
df['TotalRecovered'] = df['TotalRecovered'].astype(float)
```

```python
df['NewCases'] = df['NewCases'].str.replace(' ', '')
df['NewCases'] = df['NewCases'].astype(float)
```

```python
df = df.sort_values(by = 'TotalDeaths',ascending = False)
df.head(10)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
ax.barh(df['Country,Other'], df['TotalDeaths'], color = ["#adb0ff", "#ffb3ff", "#90d595", "#e48381", "#aafbff", "#f7bb5f", "#eafb50", "#7AB1FF", "#B5FBDD", "#F85C50"])
ax.text(0.05, 1.10, 'COUNTRIES WITH THE HIGHEST COVID19 VIRUS DEATH CASES!!!',
            transform=ax.transAxes, size=20, weight=600, ha='left', va='top')
for i, (value, name) in enumerate(zip(df['TotalDeaths'], df['Country,Other'])):
    ax.text(value, i, value, size = 15)
    ax.text(value, i-0.4, name, size = 15)
ax.text(1, 0.9, 'By PitambarMahato', transform=ax.transAxes, color='#777777', ha='right',
            bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
ax.set_yticks([])
plt.show()
fig.savefig('3.jpg')
```
